# Some text processing functions in sklearn
## CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()
corpus = ['This is the first document.',
          'This is the second second document.',
          'And the third one.',
          'Is this the first document?']
X = vectorizer.fit_transform(corpus)
X
```

```python
vectorizer.get_feature_names()
```

```python
X.toarray()
```

```python
vectorizer.transform(['Something completely new.']).toarray()
```

```python
vectorizer.transform(['A document with repeated document word']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(X)
```

# TfidfVectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
X_tfidf = vectorizer.fit_transform(corpus)
X_tfidf
```

```python
X_tfidf.toarray()
```

```python
cosine_similarity(X_tfidf)
```

## HashingVectorizer

```python
from sklearn.feature_extraction.text import HashingVectorizer
hv = HashingVectorizer(n_features=5)
hv.transform(corpus).toarray()
```

```python
cosine_similarity(hv.transform(corpus))
```

# Topic Modelling
Watch http://stat-graphics.org/movies/ldavis.html for an example of the use of pyLDAvis

```python
from sklearn.decomposition import LatentDirichletAllocation
import pyLDAvis
import pyLDAvis.sklearn
```

```python
lda_model = LatentDirichletAllocation(n_components=20, # Number of topics
                                      learning_method='online', # Faster when using large volumes of data
                                      random_state=0,       
                                      n_jobs = -1  # Use all available CPUs
                                     )
```

```python
lda_output = lda_model.fit_transform(vectors)
```

```python
pyLDAvis.enable_notebook()
pyLDAvis.sklearn.prepare(lda_model, vectors, tweets_vectorizer, mds='tsne')
```

Topic 1 is much larger than the rest. If you hover over topic 1 you will see that most of the words are stop words (which we have not removed). Let's remove stop words and repeat the analysis.

```python
tweets_vectorizer_nonstop = TfidfVectorizer(stop_words="english")
vectors_nonstop = tweets_vectorizer_nonstop.fit_transform(tweets_pd.text)
vectors_nonstop
```

```python
lda_model_nonstop = LatentDirichletAllocation(n_components=20, # Number of topics
                                              learning_method='online', # Faster when using large volumes of data
                                              random_state=0,       
                                              n_jobs = -1  # Use all available CPUs
                                              )
```

```python
lda_output_nonstop = lda_model_nonstop.fit_transform(vectors_nonstop)
```

```python
lda_vectors = pd.DataFrame(lda_output_nonstop)
lda_vectors.columns = ['Topic '+str(i) for i in range(lda_vectors.shape[1])]
lda_vectors.index = ['Document '+str(i) for i in range(lda_vectors.shape[0])]
lda_vectors.head()
```

```python
pyLDAvis.sklearn.prepare(lda_model_nonstop, vectors_nonstop, tweets_vectorizer_nonstop, mds='tsne')
```

```python
# Function from https://towardsdatascience.com/topic-modeling-quora-questions-with-lda-nmf-aff8dce5e1dd
import numpy as np
def show_topics(vectorizer=vectorizer, lda_model=lda_model, n_words=20):
    keywords = np.array(vectorizer.get_feature_names())
    topic_keywords = []
    for topic_weights in lda_model.components_:
        top_keyword_locs = (-topic_weights).argsort()[:n_words]
        topic_keywords.append(keywords.take(top_keyword_locs))
    return topic_keywords

topic_keywords = show_topics(vectorizer=tweets_vectorizer, lda_model=lda_model, n_words=20)

df_topic_keywords = pd.DataFrame(topic_keywords)
df_topic_keywords.columns = ['Word '+str(i) for i in range(df_topic_keywords.shape[1])]
df_topic_keywords.index = ['Topic '+str(i) for i in range(df_topic_keywords.shape[0])]
df_topic_keywords.T
```
